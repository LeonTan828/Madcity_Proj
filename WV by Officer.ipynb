```python
import pandas
```

```python
incidentRep = pandas.read_csv("Police_Incident_Reports.csv")
incidentRep
```

```python
# 2018
y2018 = incidentRep.loc[0:924, ['IncidentType', 'IncidentDate', 'ReleasedBy']]
y2018
```

```python
currRow = 0 

for i, row in incidentRep.iterrows():
    newstr = row["IncidentDate"][:4]
    if newstr == "2018":
        y2018.loc[currRow, "IncidentType"] = row["IncidentType"]
        y2018.loc[currRow, "IncidentDate"] = row["IncidentDate"]
        y2018.loc[currRow, "ReleasedBy"] = row["ReleasedBy"]
        currRow = currRow + 1
        
y2018
```

```python
axy2018=y2018["ReleasedBy"].value_counts().plot.bar()
axy2018
```

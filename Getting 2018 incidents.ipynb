```python
import pandas
from geopy.geocoders import ArcGIS
import time
```

```python
nom = ArcGIS()
```

```python
incidentRep = pandas.read_csv("Police_Incident_Reports.csv")
incidentRep
```

```python
newTab = incidentRep.loc[0:924, "Address"]
newTab = newTab.to_frame()
```

```python
currRow = 0 

for i, row in incidentRep.iterrows():
    newstr = row["IncidentDate"][:4]
    if newstr == "2018":
        newTab.loc[currRow, "Address"] = row["Address"]
        currRow = currRow + 1
        
```

```python
newTab
```

```python
newTab["AddrCoor"] = newTab["Address"] + ", Madison, WI"
```

```python
number = 925
for i, row in newTab.iterrows():
    if number%5 == 0:
        print(number/5)
        
    time.sleep(1)
    row["AddrCoor"] = nom.geocode(row["AddrCoor"])
    
print("DONE!!!")
```

```python
number = 925
for i, row in newTab.iterrows():
    if i < 763:
        continue
        
    print(number)
    time.sleep(1)
    row["AddrCoor"] = nom.geocode(row["AddrCoor"])
    
print("DONE!!!")
```

```python
newTab["Latitude"] = newTab["AddrCoor"].apply(lambda x: x.latitude)
```

```python
newTab["Longitude"] = newTab["AddrCoor"].apply(lambda x: x.longitude)
newTab
```

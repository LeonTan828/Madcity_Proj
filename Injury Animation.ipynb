```python
# my version

city = geopandas.read_file("city")
lakes = geopandas.read_file("lakes")



def update_func(frame_num):
    
    if frame_num == 0:
        injurypoints = geopandas.GeoDataFrame(points2008injury, geometry="loc")
        nonpoints = geopandas.GeoDataFrame(points2008non, geometry="loc")
        year = 2008
    elif frame_num == 1:
        injurypoints = geopandas.GeoDataFrame(points2009injury, geometry="loc")
        nonpoints = geopandas.GeoDataFrame(points2009non, geometry="loc")
        year = 2009
    elif frame_num == 2:
        injurypoints = geopandas.GeoDataFrame(points2010injury, geometry="loc")
        nonpoints = geopandas.GeoDataFrame(points2010non, geometry="loc")
        year = 2010
    elif frame_num == 3:
        injurypoints = geopandas.GeoDataFrame(points2011injury, geometry="loc")
        nonpoints = geopandas.GeoDataFrame(points2011non, geometry="loc")
        year = 2011
    elif frame_num == 4:
        injurypoints = geopandas.GeoDataFrame(points2012injury, geometry="loc")
        nonpoints = geopandas.GeoDataFrame(points2012non, geometry="loc")
        year = 2012
    elif frame_num == 5:
        injurypoints = geopandas.GeoDataFrame(points2013injury, geometry="loc")
        nonpoints = geopandas.GeoDataFrame(points2013non, geometry="loc")
        year = 2013
    elif frame_num == 6:
        injurypoints = geopandas.GeoDataFrame(points2014injury, geometry="loc")
        nonpoints = geopandas.GeoDataFrame(points2014non, geometry="loc")
        year = 2014
    elif frame_num == 7:
        injurypoints = geopandas.GeoDataFrame(points2015injury, geometry="loc")
        nonpoints = geopandas.GeoDataFrame(points2015non, geometry="loc")
        year = 2015
    elif frame_num == 8:
        injurypoints = geopandas.GeoDataFrame(points2016injury, geometry="loc")
        nonpoints = geopandas.GeoDataFrame(points2016non, geometry="loc")
        year = 2016
    elif frame_num == 9:
        injurypoints = geopandas.GeoDataFrame(points2017injury, geometry="loc")
        nonpoints = geopandas.GeoDataFrame(points2017non, geometry="loc")
        year = 2017
    elif frame_num == 10:
        injurypoints = geopandas.GeoDataFrame(points2018injury, geometry="loc")
        nonpoints = geopandas.GeoDataFrame(points2018non, geometry="loc")
        year = 2018

    ax.cla() # clears plot area (try removing it!)
    city.plot(color="lightgray", ax=ax)
    lakes.plot(color="darkgray", ax=ax)
    injurypoints.plot(color="red", marker="x", ax=ax)
    nonpoints.plot(color="blue", marker="x", ax=ax)
    ax.set_axis_off()
    
    ax.text(-89.384054, 43.015, year)
    ax.text(-89.384054, 43.008, "Red for violations with likely injury")
    ax.text(-89.384054, 43.001, "Blue for other violations")
    

with open("InjuryAnimation.html", "w") as f:
    f.write(animation.FuncAnimation(fig, update_func, frames=10, interval=2000).to_html5_video())
```

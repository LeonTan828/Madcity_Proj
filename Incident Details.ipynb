```python
import pandas
```

```python
incidentRep = pandas.read_csv("Police_Incident_Reports.csv")
incidentRep
```

```python
# 2018
y2018 = incidentRep.loc[0:924, ['IncidentType', 'IncidentDate', 'Details']]
y2018
```

```python
currRow = 0 

for i, row in incidentRep.iterrows():
    newstr = row["IncidentDate"][:4]
    if newstr == "2018":
        y2018.loc[currRow, "IncidentType"] = row["IncidentType"]
        y2018.loc[currRow, "IncidentDate"] = row["IncidentDate"]
        y2018.loc[currRow, "Details"] = row["Details"]
        currRow = currRow + 1
        
y2018
```

```python
# 2018 WV only
y2018wv = y2018.loc[0:215, ['IncidentType', 'IncidentDate', 'Details']]
y2018wv
```

```python
currRow = 0

for i, row in y2018.iterrows():
    newstr = row["IncidentDate"][:4]
    currInc = row['IncidentType']
    
    if currInc == "Weapons Violation":
        if newstr == "2018":
            y2018wv.loc[currRow, "IncidentType"] = row["IncidentType"]
            y2018wv.loc[currRow, "IncidentDate"] = row["IncidentDate"]
            y2018wv.loc[currRow, "Details"] = row["Details"]
            currRow = currRow + 1
            
y2018wv
```

```python
import pandas
from geopy.geocoders import ArcGIS
import time
```

```python
nom = ArcGIS()
```

```python
incidentRep = pandas.read_csv("Police_Incident_Reports.csv")
incidentRep
```

```python
# 2018
tab2018 = incidentRep.loc[0:215, "Address"]
tab2018 = tab2018.to_frame()
tab2018
```

```python
currRow = 0 

for i, row in incidentRep.iterrows():
    newstr = row["IncidentDate"][:4]
    currInc = row['IncidentType']
    
    if newstr == "2018":
        if currInc == "Weapons Violation":
            tab2018.loc[currRow, "Address"] = row["Address"]
            currRow = currRow + 1

tab2018
```

```python
tab2018["AddrCoor"] = tab2018["Address"] + ", Madison, WI"
tab2018
```

```python
for i, row in tab2018.iterrows():
    
    print(0)
        
    time.sleep(1)
    row["AddrCoor"] = nom.geocode(row["AddrCoor"])
    
print("DONE!!!")
```

```python
tab2018.iloc[63]
```

```python
tab2018.iloc[64]
```

```python
for i, row in tab2018.iterrows():
    if i < 64:
        continue
    
    print(0)
    time.sleep(1)
    row["AddrCoor"] = nom.geocode(row["AddrCoor"])
    
print("DONE!!!")
```

```python
tab2018["Latitude"] = tab2018["AddrCoor"].apply(lambda x: x.latitude)
```

```python
tab2018["Longitude"] = tab2018["AddrCoor"].apply(lambda x: x.longitude)
tab2018
```

```python
tab2018.to_csv("WeaponVio2018.csv", encoding = 'utf-8', index = False)
```
